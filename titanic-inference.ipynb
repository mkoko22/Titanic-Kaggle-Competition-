```python
import mlflow
import dagshub
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# 1. Connect to DagsHub
dagshub.init(repo_owner='mkoko22', repo_name='Titanic-Kaggle-Competition-', mlflow=True)

DATA_RUN_ID = "7c0420e1b6dc4231893610c8a7dee9d0" 

# --- 2. Download the Data ---
print("Downloading preprocessed test data...")
local_data_path = mlflow.artifacts.download_artifacts(run_id=DATA_RUN_ID, artifact_path="final_preprocessed_test.csv")
X_test = pd.read_csv(local_data_path)
print("Clean Test Data Loaded!")

# --- 3. Load the Registered Model ---
print("Loading Registered Model 'DecisionTree' (Version 1)...")
model_uri = "models:/DecisionTree/1"

try:
    # Load directly from the registry
    model = mlflow.sklearn.load_model(model_uri)
    print("Model Loaded into memory!")
    
    # --- 4. Predict ---
    print("Generating Predictions...")
    predictions = model.predict(X_test)
    
    # Grab PassengerId from original test file
    original_test_df = pd.read_csv('/kaggle/input/competitions/titanic/test.csv') 
    
    # Create submission file
    submission = pd.DataFrame({
        "PassengerId": original_test_df["PassengerId"],
        "Survived": predictions.astype(int)
    })
    
    submission.to_csv('submission.csv', index=False)
    print("SUCCESS! 'submission.csv' is ready!")
    display(submission.head())

except Exception as e:
    print(f"Error: {e}")
```
